# Here we will analyse the data by framing questions on the data set.

```python
# importing libraries.

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

## here we are creating objects for LabelEncoder and MinMaxScaler.

```python
LE = LabelEncoder()
scaler = MinMaxScaler()
```

```python
refined_data = pd.read_csv("refined_data_set.csv")
```

## Q1) What is the distribution of rating for courses ?

```python
sns.set(style='darkgrid')
sns.countplot(x='star_ratings', data=refined_data, palette='rocket')
plt.title("Bar plot for count of star rating over all.")
plt.xlabel("Star ratings.")
plt.ylabel("Frequency.")
plt.show()
```

## Q2) What is the star rating for courses according to duration? Also mention its level.

```python
sns.set(style='darkgrid')
plt.figure(figsize=(10,5))
sns.barplot(x='duration', y='star_ratings', data=refined_data, hue="level", ci=None, palette='rocket')
plt.title("Bar plot for duration and star rating along with level of course.")
plt.xlabel("duration.")
plt.ylabel("star_ratings.")
plt.show()
```

## Q3) What are the skills most in demand in accordance to the frequency of them in courses?

```python
from ast import If
s6 = 0
s7 = 0
s8 = 0
s9 = 0
s10 = 0
s11 = 0
s12 = 0
s13 = 0
s14 = 0
s15 = 0
print(refined_data.shape[0])
for i in range(refined_data.shape[0]):
  if(refined_data.iloc[i,6] ==1):
    s6+=1
  if(refined_data.iloc[i,7]==1):
    s7+=1
  if(refined_data.iloc[i,8]==1):
    s8+=1
  if(refined_data.iloc[i,9]==1):
    s9+=1
  if(refined_data.iloc[i,10]==1):
    s10+=1
  if(refined_data.iloc[i,11]==1):
    s11+=1
  if(refined_data.iloc[i,12]==1):
    s12+=1
  if(refined_data.iloc[i,13]==1):
    s13+=1
  if(refined_data.iloc[i,14]==1):
    s14+=1
  if(refined_data.iloc[i,15]==1):
    s15+=1


names = ['Writing', 'Communication', 'Business Communication', 'Leadership and Management', 'Culture', 'Planning', 'Strategy', 'Critical Thinking', 'Professional Development', 'Others']
values = [s6, s7, s8, s9, s10, s11, s12, s13, s14, s15]

skills = pd.DataFrame({'names':names, 'values':values})
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(25,5))
sns.barplot(x='names', y='values', data=skills, ci=None, palette='rocket')
plt.title("Bar plot for skills and their frequency.")
plt.xlabel("Skills.")
plt.ylabel("Frequency.")
plt.show()
```

## Q4) Plot the number of raters vs star ratings for the courses?
