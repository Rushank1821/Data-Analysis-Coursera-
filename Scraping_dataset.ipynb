# Data Scraping for Courses on Language Learning (Coursera)

```python
# Importing Libraries

import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time
```

```python
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)
```

```python
coursera_url =f'https://www.coursera.org/search?query=language%20learning&topic=Language%20Learning&sortBy=BEST_MATCH'
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

driver.get(coursera_url)
driver.maximize_window()

#declaring list to store data.
institution_name = []
title_of_course = []
star_ratings = []
number_of_raters = []
skills_list = []
level_of_course = []

#when the show more option comes.
for i in range(1, 212):
    print("----------------------------------------------------------------------------------")
    print(f"i : {i}")

    # institution name
    try:
        company = driver.find_element(By.XPATH, f'/html/body/div[2]/div/div/main/div[2]/div/div/div/div/div[2]/ul/li[{i}]/div/div/div/div/div/div[2]/div[1]/div[1]/div/div/p').text
        print(f"company: {company}")
    except:
        company = ""
        print(company)

    institution_name.append(company);

    # title of the course
    try:
        title = driver.find_element(By.XPATH, f'/html/body/div[2]/div/div/main/div[2]/div/div/div/div/div[2]/ul/li[{i}]/div/div/div/div/div/div[2]/div[1]/div[2]/a/h3').text
        print(f"title : {title}")
    except:
        title = ""
        print(title)

    title_of_course.append(title)

    #star rating
    try:
        star = driver.find_element(By.XPATH, f'/html/body/div[2]/div/div/main/div[2]/div/div/div/div/div[2]/ul/li[{i}]/div/div/div/div/div/div[2]/div[3]/div[2]/div/div/p[1]').text
        print(f"star rating : {star}")
    except:
        try:
            star = driver.find_element(By.XPATH, f'/html/body/div[2]/div/div/main/div[2]/div/div/div/div/div[2]/ul/li[{i}]/div/div/div/div/div/div[2]/div[2]/div[2]/div/div/p[1]').text
            print(f"star rating : {star}")
        except:
            star = ""
            print(star)

    star_ratings.append(star)

    #number of ratings
    try:
        no_of_ratings = driver.find_element(By.XPATH, f'/html/body/div[2]/div/div/main/div[2]/div/div/div/div/div[2]/ul/li[{i}]/div/div/div/div/div/div[2]/div[3]/div[2]/div/div/p[2]').text
        print(f"number of ratings : {no_of_ratings}")
    except:
        try:
            no_of_ratings = driver.find_element(By.XPATH, f'/html/body/div[2]/div/div/main/div[2]/div/div/div/div/div[2]/ul/li[{i}]/div/div/div/div/div/div[2]/div[2]/div[2]/div/div/p[2]').text
            print(f"number of ratings : {no_of_ratings}")
        except:
            no_of_ratings = ""
            print(no_of_ratings)

    number_of_raters.append(no_of_ratings)

    #skills
    try:
        skills = driver.find_element(By.XPATH, f'/html/body/div[2]/div/div/main/div[2]/div/div/div/div/div[2]/ul/li[{i}]/div/div/div/div/div/div[2]/div[2]/div/p').text
        if(skills[0:19]=="Skills you'll gain:"):
            print(skills)
            skills_list.append(skills)
        else:
            skills_list.append("")
            print("")
    except:
        print("")

    #level of course(optional)
    try:
        level = driver.find_element(By.XPATH, f'/html/body/div[2]/div/div/main/div[2]/div/div/div/div/div[2]/ul/li[{i}]/div/div/div/div/div/div[2]/div[3]/div[3]/p').text
        print(f"level : {level}")
    except:
        try:
            level = driver.find_element(By.XPATH, f'/html/body/div[2]/div/div/main/div[2]/div/div/div/div/div[2]/ul/li[{i}]/div/div/div/div/div/div[2]/div[2]/div[3]/p').text
            print(f"level : {level}")
        except:
            level = ""
            print("")
    level_of_course.append(level)


    if(i%12==0):
        show_more = driver.find_element(By.XPATH, '/html/body/div[2]/div/div/main/div[2]/div/div/div/div/div[2]/div[4]/button').click()
        # print("show more")
        time.sleep(5)
```
